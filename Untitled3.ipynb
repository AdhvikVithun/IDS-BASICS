```python
# Initialize an empty dictionary
books = {}

# Taking input from the user to fill the dictionary
n = int(input("Enter the number of books you want to add: "))

# Adding book_id and book_name to the dictionary
for i in range(n):
    book_id = input(f"Enter book ID for book {i + 1}: ")
    book_name = input(f"Enter book name for book {i + 1}: ")
    books[book_id] = book_name

# Taking input for a book_id to search
search_id = input("\nEnter the book ID you want to search for: ")

# Traversing the dictionary to find the book_name
if search_id in books:
    print(f"The book name for ID '{search_id}' is: {books[search_id]}")
else:
    print("Book ID not found.")
```
